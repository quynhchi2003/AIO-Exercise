```python
def compute_average(vector):
    return vector.mean()
```

```python
def lightness(vector):
    return (vector.max() + vector.min())/2
```

```python
def luminosity(vector):
    return vector[0]*0.21 + vector[1]*0.72 + vector[2]*0.07
```

```python
img = mpimg.imread('./dog.jpeg')
img =  np.apply_along_axis(compute_average, axis = 2, arr=img)
plt.imshow(img, cmap='gray')
img[0,0]
```

```python
img = mpimg.imread('./dog.jpeg')
img =  np.apply_along_axis(lightness, axis = 2, arr=img)
plt.imshow(img, cmap='gray')
img[0,0]
```

```python
img = mpimg.imread('./dog.jpeg')
img =  np.apply_along_axis(luminosity, axis = 2, arr=img)
plt.imshow(img, cmap='gray')
img[0,0]
```

```python
df = pd.read_csv('./advertising.csv')
df.head()
```

```python
data = df.to_numpy()
```

```python
print(data[:,3].argmax())
print(data[:,3].max())
```

```python
print(data[:,0].mean())
```

```python
print(sum(data[:,3]>=20))
```

```python
filtered_data = data[data[:, 3] >= 15]
mean_radio = np.mean(filtered_data[:, 1])
print(mean_radio)
```

```python
filtered_data = data[data[:, 2] >= data[:, 2].mean()]
total_sale = np.sum(filtered_data[:, 3])
print(total_sale)
```

```python
mean_sales = np.mean(data[:, 3])
scores = np.array(["Good" if sales > mean_sales else "Average" if sales == mean_sales else "Bad" for sales in data[:, 3]])
print(scores[7:10])
```

```python
mean_sales = np.mean(data[:, 3])
closest_to_mean = data[:, 3][np.abs(data[:, 3] - mean_sales).argmin()]
scores = np.array(["Good" if sales > closest_to_mean else "Average" if sales == closest_to_mean else "Bad" for sales in data[:, 3]])
print(scores[7:10])
```
